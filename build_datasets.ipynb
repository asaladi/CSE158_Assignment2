```python
import random
import os
import loader 
import pandas as pd
from collections import defaultdict
```

```python
root = "datasets2/"
train_interactions = pd.read_csv(root+"interactions_train.csv")
val_interactions = pd.read_csv(root+"interactions_validation.csv")
test_interactions = pd.read_csv(root+"interactions_test.csv")
```

```python
all_interactions = pd.concat([train_interactions, val_interactions, test_interactions], axis=0, ignore_index=True)

all_interactions = all_interactions.sample(frac=1, random_state=42).reset_index(drop=True)
all_interactions = all_interactions.head(100000).reset_index(drop=True)
print(len(all_interactions))
```

```python
hidden_frac = 0.05
hidden_interactions = all_interactions.sample(frac=hidden_frac, random_state=42)
seen_interactions = all_interactions.drop(hidden_interactions.index)
```

```python
output_root = "datasets3/"
os.makedirs(output_root, exist_ok=True)

# Save CSVs
hidden_interactions.to_csv(output_root + "hidden_interactions.csv", index=False)
seen_interactions.to_csv(output_root + "seen_interactions.csv", index=False)

print("Hidden interactions shape:", hidden_interactions.shape)
print("Seen interactions shape:", seen_interactions.shape)
print(f"Saved to {output_root}")
```
